```python
import pandas as pd
data = pd.read_csv('Car details v3.csv', sep = ',')
data.head()
```

```python
pd.value_counts(data['owner'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,6))
plt.bar(data['owner'].unique(),data['owner'].value_counts())
plt.title('Propietario')
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.hist(data['selling_price'], bins=20)
plt.title('Precio')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(data['km_driven'],data['selling_price'],'.',color='blue')
plt.title('Km recorridos vs Precio')
plt.xlabel('Km recorridos')
plt.ylabel('Precio')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(16,6))
plt.tight_layout()
sns.distplot(data['selling_price'])
```
